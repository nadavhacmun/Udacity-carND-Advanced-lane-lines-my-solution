# Camera calibration
* Read in calibration images
* Detect edges
* Create list of imgpoints and objpoints
* Use cv2.calibrateCamera() to find camera matrix and distortion coefficients
* Undistorting an image and plotting as an example

```python
%matplotlib inline

objpoints = [] # 3D points in real world 
imgpoints = [] # 2D points in image 
shape = None # variable to store the shape of the image

objp = np.zeros((6*9, 3), np.float32) # initializing empty array

objp[:, :2] = np.mgrid[0:9, 0:6].T.reshape(-1, 2) # adding real points to it

for path in glob.glob("camera_cal/calibration*.jpg"):
    img = mpimg.imread(path) # read in RGB image
    
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) # convert RGB img to grayscale
    shape = gray.shape
    
    ret, corners = cv2.findChessboardCorners(gray, (9, 6), None) # find chessboard corners in gray image
    
    # if chessboard corners detected
    if ret:
        imgpoints.append(corners) # append found img points to imgpoints
        objpoints.append(objp) # append objp to objpoints

ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, shape[::-1], None, None)

# un-distorting image to show calibration is correct
img = mpimg.imread("camera_cal/calibration1.jpg") # read in image

undistort = cv2.undistort(img, mtx, dist, None, mtx) # un-distorting image

# plotting original image next to calibrated image
fig = plt.figure(figsize=(15, 5))

fig.add_subplot(1, 2, 1) 
plt.title("Original")
plt.axis("off") # turning off ticks
plt.imshow(img)

fig.add_subplot(1, 2, 2)
plt.title("Calibrated")
plt.axis("off") # turning off ticks
plt.imshow(undistort)

plt.show()
```

# Extracting edges
* Thresholding R and S channels
* Combining combining thresholds
* visuallizing

```python
# defining some helpful functions

def thresh(img, min_thresh, max_thresh):
    binary_output = np.zeros_like(img)
    binary_output[(img > min_thresh) & (img <= max_thresh)] = 1
    
    return binary_output

def mask_mag(img, min_thresh, max_thresh, ksize):
    xsobel = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=ksize)
    ysobel = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=ksize)
    
    mag = np.sqrt(np.square(xsobel) + np.square(ysobel))
    mag_norm = np.uint8(mag*255/np.max(mag))
    
    binary_out = np.zeros_like(mag_norm)
    binary_out[(mag_norm > min_thresh) & (mag_norm < max_thresh)] = 1
    
    return binary_out

def mask_dir(img, min_thresh, max_thresh):
    xsobel = np.absolute(cv2.Sobel(img, cv2.CV_64F, 1, 0))
    ysobel = np.absolute(cv2.Sobel(img, cv2.CV_64F, 0, 1))
    
    dir_ = np.arctan2(ysobel, xsobel)
    
    binary_out = np.zeros_like(dir_)
    binary_out[(dir_ > min_thresh) & (dir_ < max_thresh)] = 1
    
    return binary_out
```

```python
def getEdges(img):
    
    R = img[:, :, 0] # getting red color channel 

    HLS = cv2.cvtColor(np.copy(img), cv2.COLOR_RGB2HLS)
    S = HLS[:,:, 2] # getting S channel out of HLS image

    gray = cv2.cvtColor(np.copy(img), cv2.COLOR_RGB2GRAY)

    binary_mag = mask_mag(gray, 75, 255, 5) # thresh magnitude 
    binary_r = thresh(R, 200, 255) # thresh R channel from 200 to 255
    binary_s = thresh(S, 90, 255) # thresh S channel from 90 to 255

    # making a binary image 
    out = np.zeros_like(binary_r) 
    out[((binary_r == 1) & (binary_s == 1)) | (binary_mag == 1)] = 1
    
    return out, binary_r, binary_s
```

```python
img = mpimg.imread("test_images/test1.jpg") # read in image 
img = undistort(img, mtx, dist)
out, binary_r, binary_s = getEdges(img) # get edges 

fig = plt.figure(figsize=(24,15))

# constructing image with individual channels seen
visual = np.stack([np.zeros_like(out), binary_s, binary_r], axis=2)
visual *= 255

# plotting
fig.add_subplot(1, 3, 1)
plt.title("Original image")
plt.axis("off")
plt.imshow(img, cmap="gray")

fig.add_subplot(1, 3, 2)
plt.title("Edges")
plt.axis("off")
plt.imshow(out, cmap="gray")

fig.add_subplot(1, 3, 3)
plt.title("Individual channels seen")
plt.axis("off")
plt.imshow(visual)

plt.show()
```

# Perspective warping
* create transform matrix using cv2.getPerspectiveTransform
* apply to image as an example

```python
# helper function, draws quad to the image
def draw_quad(left_u, left_d, right_u, right_d, img):
    cv2.line(img, left_u, right_u, (255, 0, 0), thickness=1)
    cv2.line(img, left_u, left_d, (255, 0, 0), thickness=1)
    cv2.line(img, left_d, right_d, (255, 0, 0), thickness=1)
    cv2.line(img, right_d, right_u, (255, 0, 0), thickness=1)
```

```python
def getTransforms(img):
    """
    img: binary edges image
    """
    # original points in the image
    src = np.float32([[img.shape[1] - 600, 450],
                      [img.shape[1] - 170, img.shape[0]],
                      [170, img.shape[0]],
                      [600, 450]])

    # their destination in the warped image
    dst = np.float32([[img.shape[1] - 300, 0],
                      [img.shape[1] - 300, img.shape[0]],
                      [300, img.shape[0]],
                      [300, 0]])

    # getting the transform and inverse transform
    M = cv2.getPerspectiveTransform(src, dst)
    Minv = cv2.getPerspectiveTransform(dst, src)
    
    return M, Minv

def Warp(img, M):
    
    im_size = (img.shape[1], img.shape[0])
    
    # applying the transform to the image
    warped = cv2.warpPerspective(img, M, im_size, flags=cv2.INTER_LINEAR)
    
    return warped
```

```python
def find_lane_pixels(warped, rect=False):
    hist = np.sum(warped[np.int(warped.shape[0]/2):,:], axis=0) # sum along vertical axis over bottom half of the image

    middle = np.int(warped.shape[1] / 2) # middle X axis
    left_base = np.argmax(hist[:middle]) # getting the base for the first left window
    right_base = middle + np.argmax(hist[middle:]) # getting the base for the first right window

    bases = [left_base, right_base] # keeping bases in a list for use later in computing car position
    
    nwindows = 9 # amount of windows
    margin = 100 # X margin in every direction of the windows
    minpix = 50 # minimal amount of pixels to recenter the next window

    window_height = np.int(warped.shape[0] / nwindows) # finding window height
    nonzero = warped.nonzero() # getting nonzero pixels locations
    nonzerox = np.array(nonzero[1]) # X coordinates of nonzero pixels
    nonzeroy = np.array(nonzero[0]) # Y coordinates of nonzero pixels

    leftx_current = left_base # current center of left window
    rightx_current = right_base # current center of the right window

    left_lane_inds = [] # indices of points on the left lane
    right_lane_inds = [] # indices of the points on the right lane
    
    # making an image to plot
    img = np.dstack([warped,warped,warped])
    # go over all windows
    for window in range(nwindows):
        # get window boundries
        y_high = warped.shape[0] - window*window_height
        y_low = warped.shape[0] - (window+1)*window_height

        leftX_low = leftx_current - margin
        leftX_high = leftx_current + margin

        rightX_low = rightx_current - margin
        rightX_high = rightx_current + margin
        
        if rect:
            # draw windows
            cv2.rectangle(img, (leftX_low, y_low), (leftX_high, y_high), (0, 1.0, 0), 2)
            cv2.rectangle(img, (rightX_low, y_low), (rightX_high, y_high), (0, 1.0, 0), 2)
        
        # get indices for lanes in the windows
        good_left_inds = ((nonzerox > leftX_low) & (nonzerox < leftX_high) & (nonzeroy >= y_low) & (nonzeroy < y_high)).nonzero()[0]
        good_right_inds = ((nonzerox > rightX_low) & (nonzerox < rightX_high) & (nonzeroy >= y_low) & (nonzeroy < y_high)).nonzero()[0]

        # append them to left and right lane indices
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)

        # if more then minpix pixels are activated in the windows recenter them
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)
    
    # coordinates of points on left and right lanes
    left_x = nonzerox[left_lane_inds]
    left_y = nonzeroy[left_lane_inds]
    right_x = nonzerox[right_lane_inds]
    right_y = nonzeroy[right_lane_inds]
    
    return left_x, left_y, right_x, right_y, img, bases
    
def fit_poly(left_x, left_y, right_x, right_y, img):
    
    # fit 2nd degree polynomials to both lanes
    left_fit = np.polyfit(left_y, left_x, 2)
    right_fit = np.polyfit(right_y, right_x, 2)
            
#    else:
 #       ym_per_pix = 30/720 # meters per pixel in y dimension
  #      xm_per_pix = 3.7/700 # meters per pixel in x dimension
        
        # fit 2nd degree polynomials to both lanes
   #     left_fit = np.polyfit(left_y*ym_per_pix, left_x*xm_per_pix, 2)
    #    right_fit = np.polyfit(right_y*ym_per_pix, right_x*xm_per_pix, 2)
        
    
    # domain for the functions we fitted
    ploty = np.linspace(0, img.shape[0]-1, img.shape[0])
    
    # range for the functions we fitted
    leftX_cords = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    rightX_cords = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    
    # color in left and right lane
    img[left_y, left_x] = [1.0,0.0, 0.0]
    img[right_y, right_x] = [0.0, 0.0, 1.0]
    
    return img, leftX_cords, rightX_cords, ploty, left_fit, right_fit
```

```python
# calling above functions to use
left_x, left_y, right_x, right_y, immg, bases = find_lane_pixels(warped)
immg, leftpolyX, rightpolyX, ploty, left_fit, right_fit = fit_poly(left_x, left_y, right_x, right_y, immg)

# plotting
plt.figure(figsize=(15, 15))
plt.imshow(immg*255)
plt.plot(leftpolyX, ploty)
plt.plot(rightpolyX, ploty)
# x limit since polynomial goes outside bound of the picture
plt.xlim(0, immg.shape[1])
```

```python
def find_mid(img, warped, leftpolyX, rightpolyX):
    imgg = warped
    X, Y = np.meshgrid(range(1280), range(720)) # makes a 720x1280 grid (this is the same shape as the image)

    arr = np.zeros_like(X) # makes an array full of zeros with the same shape as X 
    for i in range(imgg.shape[1] - 1): # iterate over all columns
        arr[:, i] = leftpolyX#leftX_cords # setting every column in the array to be the same as the X coordinates of the polynomial for the right lane
        #arr[:451, :] = 0 # set top 450 pixels of image back to 0 since we only detect the lane lines up to that point

    arr1 = np.zeros_like(X) # makes an array full of zeros with the same shape as X 
    for i in range(imgg.shape[1] - 1): # iterate over all columns
        arr1[:, i] = rightpolyX#rightX_cords # setting every column in the array to be the same as the X coordinates of the polynomial for the right lane
        #arr1[:451, :] = 0 # set top 450 pixels of image back to 0 since we only detect the lane lines up to that point

    locations = (X > arr) & (X < arr1) # get all locations such that their X coordinate is greater then the left polynomials X coordinate and smaller then the right one
    img[locations, 1] = 1 # set their green channel to be 1
    
    return img
```

# Calculating curvature and finding where the car is located with respect to the center of the lanes
* finding curvature at bottom of image (in meters not in pixel space)
* calculating center of lane lines
* calculating difference center of lane lines to center of image

```python
#def getCurveatureAndPosition(left_x, left_y, right_x, right_y, immg, bases):
def getCurveatureAndPosition(left_fit, right_fit, ploty, immg, bases):
    """
    
    """
    #immg, leftpolyX, rightpolyX, ploty, left_fit_scaled, right_fit_scaled = fit_poly(left_x, left_y, right_x, right_y, immg, True)
    
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/1280 # meters per pixel in x dimension

    y_eval = np.max(ploty) # getting lowest Y to evaluate curvature at the bottom of the image

    left_curverad = ((1 + (2*left_fit[0]*y_eval*ym_per_pix + left_fit[1])**2)**1.5) / np.absolute(2*left_fit[0])  # calculation of the left line 
    right_curverad = ((1 + (2*right_fit[0]*y_eval*ym_per_pix + right_fit[1])**2)**1.5) / np.absolute(2*right_fit[0])  # calculation of the right line 
    
    # reusing some code from the find_lane_pixels function to get base lane coordinates
    lanes_mean = sum(bases)/2.0 # taking the mean of the bases of the lane lines
    middle = immg.shape[1] / 2.0 # middle of X axis
    car_location = xm_per_pix*(middle - lanes_mean) # scaling result to real size
    
    return min(left_curverad, right_curverad), car_location
```

# Warping image back filling entire lane and adding text
* using the inverse transform to return image to normal
* color in all pixels that are a part of the lane
* add curvature and vehicle writings at top of image

```python
def colorOriginal(img, coloredWarped, Minv):
    
    im_size = (warped.shape[1], warped.shape[0]) # cv2 needs shape to be like this for some reason
    img = img/255.0
    
    unwarped = cv2.warpPerspective(coloredWarped, Minv, im_size, flags=cv2.INTER_LINEAR) # unwarping colored lane lines image
    
    nonzero = unwarped.nonzero()
    nonzeroX = nonzero[1]
    nonzeroY = nonzero[0]
    
    img[nonzeroY, nonzeroX, :] = unwarped[nonzeroY, nonzeroX, :]
    
    return img
```

```python
def AddTextToImg(img, curvature, car_location):
    font = ImageFont.truetype("arial.ttf", size=30) # setting font type
    immg = Image.fromarray(np.uint8(img * 255)) # converting np array to Image object to add text to it
    draw = ImageDraw.Draw(immg) 
    draw.text(xy=(50,50), text="Radius of curve: " + str(int(curvature))+ "(M)", font=font) # draw the radius of the curve

    if car_location <= 0: 
        draw.text(xy=(50,100), text="car is {0:.3f}(M) to the left of center".format(-1*car_location), font=font) # text in the case when car is to the right

    else: 
        draw.text(xy=(50,100), text="car is {0:.3f}(M) to the right of center".format(car_location), font=font) # text in the case when car is to the left
        
    return immg
```
